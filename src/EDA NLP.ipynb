-  Import Libraries & SetUp

```python
import re
from sentence_transformers import SentenceTransformer, util

print("BERT Loading")
model = SentenceTransformer('all-MiniLM-L6-v2')
```

- Cleaning

```python
def clean_text(text):
    if not isinstance(text, str): return ""
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

- The Matching Engine

```python
def calculate_match(resume_text, job_desc_text):
    resume_clean = clean_text(resume_text)
    jd_clean = clean_text(job_desc_text)
    
    resume_vec = model.encode(resume_clean, convert_to_tensor=True)
    jd_vec = model.encode(jd_clean, convert_to_tensor=True)
    
    score = util.cos_sim(resume_vec, jd_vec).item() * 100
    return score
```
